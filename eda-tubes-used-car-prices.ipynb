### Cek Data yang Kosong

```python
def null_count(df):
    return pd.DataFrame({
        'Kolom': df.columns,
        'Jumlah NaN': df.isnull().sum().values,
        'Persentase NaN': df.isnull().sum().values / df.shape[0]
    }).style.background_gradient(cmap='Set2', low=0.1, high=0.01)
```

### Konversi Beberapa Kolom Non-Numerik ke Kolom Numerik

#### Australia

```python
columns_to_check = ['Price', 'Kilometres', 'Year', 'CylindersinEngine', 'FuelConsumption']

non_numeric_values = {col: [] for col in columns_to_check}

for column in columns_to_check:
    values = list(df_aus[column])
    for val in values:
        try:
            float(val)
        except:
            non_numeric_values[column].append(val)

for column, non_numeric in non_numeric_values.items():
    if non_numeric:
        print(f"Nilai non-numerikal di kolom '{column}':")
        for val in non_numeric:
            print(val, end=" ")
        print("\n")
```

```python
df_aus[['Kilometres', 'Year', 'CylindersinEngine', 'FuelConsumption']] = df_aus[['Kilometres', 'Year', 'CylindersinEngine', 'FuelConsumption']].replace('-', np.nan)
df_aus['Price'] = df_aus['Price'].replace('POA', np.nan)
df_aus.dropna(inplace=True)

df_aus['Engine'] = df_aus['Engine'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df_aus[['Year', 'Price']] = df_aus[['Year', 'Price']].astype(float)
df_aus['CylindersinEngine'] = df_aus['CylindersinEngine'].str.extract(r'(\d+)').astype(float)
df_aus['FuelConsumption'] = df_aus['FuelConsumption'].str.extract(r'(\d+\.?\d*)').astype(float)
df_aus['Kilometres'] = df_aus['Kilometres'].astype(float)

new_column_names = {
    'Price': 'Price (AU$)',
    'Kilometres': 'Mileage (km)',
    'Engine': 'Engine Size (L)',
    'CylindersinEngine': 'Cylinders in Engine (cyl)',
    'FuelConsumption': 'Fuel Consumption (L/100 km)',
    'UsedOrNew': 'Used/New',
    'DriveType': 'Drive Type',
    'BodyType': 'Body Type',
    'FuelType': 'Fuel Type',
}

df_aus.rename(columns=new_column_names, inplace=True)
df_aus.rename(columns=lambda x: x.title(), inplace=True)
```

#### Britania Raya

```python
df_uk['Engine'] = df_uk['Engine'].str.rstrip('L').astype(float)
df_uk['Engine'].fillna(0, inplace=True)
df_uk['Mileage(miles)'] = df_uk['Mileage(miles)'] * 1.60934
df_uk['Emission Class'] = df_uk['Emission Class'].str.replace('Euro', '').astype(float)

new_column_names = {
    'Mileage(miles)': 'Mileage (km)',
    'Price': 'Price ()',
    'Engine': 'Engine Size (L)',
    'Registration_Year': 'Year'
}


df_uk.rename(columns=new_column_names, inplace=True)
def capitalize_column_name(column_name):
    parts = column_name.split(' ')
    capitalized_parts = [part.capitalize() if '(' not in part else part for part in parts]
    return ' '.join(capitalized_parts)

df_uk.rename(columns=capitalize_column_name, inplace=True)
```

#### Indonesia

```python
df_id['Engine Size (L)'] = df_id['car name'].str.extract(r'(\d+\.\d+|\d+)').astype(float)

new_column_names = {
    'mileage (km)': 'Mileage (km)',
    'price (Rp)': 'Price (IDR)',
    'instalment (Rp|Monthly)': 'Installment (IDR/Month)'
}

df_id.rename(columns=new_column_names, inplace=True)
def capitalize_column_name(column_name):
    parts = column_name.split(' ')
    capitalized_parts = [part.capitalize() if '(' not in part else part for part in parts]
    return ' '.join(capitalized_parts)

df_id.rename(columns=capitalize_column_name, inplace=True)
```

### Definisikan Kolom Numerik dan Kolom Kategorikal

#### Australia

```python
aus_numerical_columns = []
aus_categorical_columns = []

for column in df_aus.columns:
    if pd.api.types.is_numeric_dtype(df_aus[column].dtype):
        aus_numerical_columns.append(column)
    else:
        aus_categorical_columns.append(column)
        
print("Kolom Numerikal:", aus_numerical_columns)
print("Kolom Kategorikal:", aus_categorical_columns)
```

#### Britania Raya

```python
uk_numerical_columns = []
uk_categorical_columns = []

for column in df_uk.columns:
    if pd.api.types.is_numeric_dtype(df_uk[column].dtype):
        uk_numerical_columns.append(column)
    else:
        uk_categorical_columns.append(column)
        
print("Kolom Numerikal:", uk_numerical_columns)
print("Kolom Kategorikal:", uk_categorical_columns)
```

#### Indonesia

```python
id_numerical_columns = []
id_categorical_columns = []

for column in df_id.columns:
    if pd.api.types.is_numeric_dtype(df_id[column].dtype):
        id_numerical_columns.append(column)
    else:
        id_categorical_columns.append(column)
        
print("Kolom Numerikal:", id_numerical_columns)
print("Kolom Kategorikal:", id_categorical_columns)
```
